```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([transforms.ToTensor()])
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('./data', train=True, download=True, transform=transform),
    batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('./data', train=False, transform=transform),
    batch_size=1000, shuffle=False)
```

```python
class SparseLinear(nn.Module):
    def __init__(self, in_features, out_features, sparsity=0.8):
        super(SparseLinear, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.sparsity = sparsity

        # Initialize weights and mask
        self.weight = nn.Parameter(torch.randn(out_features, in_features) * 0.1)
        self.mask = torch.rand_like(self.weight) > sparsity  # Random sparse mask
        self.mask = self.mask.float()

        self.bias = nn.Parameter(torch.zeros(out_features))

    def forward(self, x):
        # Apply mask to simulate sparse connections
        w = self.weight * self.mask
        return F.linear(x, w, self.bias)

    def prune_and_regrow(self):
        # Prune: Set smallest weights to 0
        with torch.no_grad():
            weight_data = (self.weight * self.mask).abs()
            threshold = torch.quantile(weight_data[weight_data > 0], 0.2)
            prune_mask = (weight_data > threshold).float()
            self.mask *= prune_mask

            # Regrow: Add new connections randomly
            num_to_regrow = int((self.mask == 0).sum().item() * 0.2)
            new_mask = torch.zeros_like(self.mask).bool()
            indices = torch.nonzero(self.mask == 0)
            chosen = indices[torch.randperm(len(indices))[:num_to_regrow]]
            new_mask[chosen[:, 0], chosen[:, 1]] = 1
            self.mask = torch.clamp(self.mask + new_mask.float(), 0, 1)
```

```python
class SparseNet(nn.Module):
    def __init__(self):
        super(SparseNet, self).__init__()
        self.fc1 = SparseLinear(28*28, 300, sparsity=0.8)
        self.fc2 = SparseLinear(300, 100, sparsity=0.8)
        self.fc3 = SparseLinear(100, 10, sparsity=0.8)

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)

    def evolve(self):
        self.fc1.prune_and_regrow()
        self.fc2.prune_and_regrow()
        self.fc3.prune_and_regrow()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SparseNet().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

def train(model, loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

    # Evolve sparse structure every 3 epochs
    if epoch % 3 == 0:
        model.evolve()

def test(model, loader):
    model.eval()
    correct = 0
    with torch.no_grad():
        for data, target in loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            pred = output.argmax(dim=1)
            correct += pred.eq(target).sum().item()
    return correct / len(loader.dataset)

# Training loop
for epoch in range(1, 11):
    train(model, train_loader, optimizer, epoch)
    acc = test(model, test_loader)
    print(f"Epoch {epoch} - Test Accuracy: {acc:.4f}")
```

```python
def calc_sparsity(layer):
    total = layer.mask.numel()
    active = layer.mask.sum().item()
    return 100 * active / total

print("Sparsity in fc1:", calc_sparsity(model.fc1), "%")
print("Sparsity in fc2:", calc_sparsity(model.fc2), "%")
print("Sparsity in fc3:", calc_sparsity(model.fc3), "%")
```
